```python
dict3 = {'College':['JNTUA','JNTUK',np.NaN,'JNTUK'],
         'Strength':[5000, np.NaN, 2500, 3890], 
         'Year':[np.NaN, 50, 40, 60],
         'Placed':[4000,3000,2000,np.NaN]}
df2 = pd.DataFrame(dict3)
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
# If column is Object type, relpace null value with,
   # Most occured value
# If 2nd col contains null value, replace with before value
# If 3rd col contains null value, replace with next value
# for Remaining null values. replace with '0' 
```

```python
df2['College'].value_counts()
```

```python
df2['College'].fillna('JNTUK', inplace = True)
df2['Strength'].fillna(method = 'ffill', inplace = True)
df2['Year'].fillna(method = 'bfill', inplace = True)
```

## Machine Learning

- Definition: Machine can learn from history or past data and make predictions for new data points.
- In other way make machines to think like a human.
- Supervised learning
    - Machine learns under supervision.
    - Two types:
        - Classification
            - We have to just classify the things like male or female, o or 1, spam or not spam etc,.
        - Regression
            - house price prediction, temperature prediction, student grade prediction.
            
- Unsupervised learning
- Reinforcement Learning

- data  --> raw format
- information --> extracting the needful data
    - with labels
        - feature columns / input columns
        - target columns / output columns
    - Without labels
        - only inputs columns
